```python
# STEP1_download_data.py

import yfinance as yf
import pandas as pd

# ✅ 19 Indian Property/Realty Sector Stocks
stocks = [
    "DLF.NS", "LODHA.NS", "GODREJPROP.NS", "SOBHA.NS", "BRIGADE.NS",
    "SUNTECK.NS", "PHOENIXLTD.NS", "OBEROIRLTY.NS", "PRESTIGE.NS",
    "MAHLIFE.NS", "ANANTRAJ.NS", "KOLTEPATIL.NS",
    "SIGNATURE.NS", "DBREALTY.NS", "ARVSMART.NS",
    "SHRIRAMPPS.NS", "RUSTOMJEE.NS", "PENINLAND.NS",
    "IBREALEST.NS"
]

all_data = []

for stock in stocks:
    print(f"Downloading {stock}...")
    df = yf.download(
        stock,
        start="2020-01-01",
        end="2025-07-31",
        auto_adjust=False,
        progress=False
    )

    if df.empty:
        print(f"⚠ No data for {stock}, skipping...")
        continue

    # Fix MultiIndex columns if present
    if isinstance(df.columns, pd.MultiIndex):
        df.columns = df.columns.get_level_values(0)

    df = df[["Open", "High", "Low", "Close", "Volume"]].copy()
    df["Ticker"] = stock
    df.reset_index(inplace=True)
    all_data.append(df)

final_df = pd.concat(all_data, ignore_index=True)

# Rename index column to Date if needed
if "index" in final_df.columns:
    final_df.rename(columns={"index": "Date"}, inplace=True)

final_df.to_csv("indian_realty_clean_data.csv", index=False)
print(f"\n✅ Data saved! Shape: {final_df.shape}")
print(final_df.head())
```
